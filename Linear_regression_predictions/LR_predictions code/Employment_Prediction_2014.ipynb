```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
employment_df = pd.read_csv('employment_clean.csv')
employment_df
```

```python
df_2019 = employment_df['2019'].copy()
df_2019
```

```python

employment_df.drop(columns = ['2019'])
```

```python
predictions = []
for i in range(0,384):
    x = [[2002],[2003],[2004],[2005],[2006],[2007],[2008],[2009],[2010],[2011],[2012],[2013],[2014],[2015],[2016],[2017],[2018]]
    y = employment_df.iloc[i, 2:].values
    model.fit(x,y)
    y_pred = model.predict([[2019]])
    predictions.append(y_pred)
flat_list = [item for sublist in predictions for item in sublist]
employment_df['prediction_2019'] = flat_list
employment_df
```

```python
employment_df.to_csv('employment_prediction_2019.csv')
```
