```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
employment_df = pd.read_csv('../Resources/Data/employment_MSA_clean.csv')
employment_df
```

```python
employment_2019_df= employment_df[["2019"]]
employment_2019_df
```

```python
employment_df=employment_df.drop(columns=["2019"], axis=1)
employment_df
```

```python
predictions = []
for i in range(0,384):
    x = [[2002],[2003],[2004],[2005],[2006],[2007],[2008],[2009],[2010],[2011],[2012],[2013],[2014],[2015],[2016],[2017],[2018]]
    y = employment_df.iloc[i, 2:].values
    model.fit(x,y)
    y_pred = model.predict([[2019]])
    predictions.append(y_pred)
flat_list = [item for sublist in predictions for item in sublist]
employment_df['prediction_2019'] = flat_list
employment_df
```

```python
compare_df = pd.concat([employment_df, employment_2019_df], axis=1)
compare_df
```

```python
compare_df['Rate_of_change'] = ((compare_df['prediction_2019']-compare_df['2019'])/compare_df['2019'])*100
compare_df
```

```python
rate_of_change_average = compare_df['Rate_of_change'].mean()
rate_of_change_average
```

```python
actual_2019_df = compare_df['2019'].sum()
actual_2019_df
```

```python
pred_2019_df = compare_df['prediction_2019'].sum()
pred_2019_df
```

```python
Diff_2019_df = ((pred_2019_df)-(actual_2019_df))/actual_2019_df
```
