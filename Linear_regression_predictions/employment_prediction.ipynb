```python
# instantiate linearRegression
model = LinearRegression()
```

```python
df
```

```python
predictions = []
for i in range(0,384):
    x = [[2001],[2002],[2003],[2004],[2005],[2006],[2007],[2008],[2009],[2010],[2011],[2012],[2013],[2014],[2015],[2016],[2017],[2018],[2019]]
    y = df.iloc[i, 1:].values
    model.fit(x,y)
    y_pred = model.predict([[2024]])
    predictions.append(y_pred)
flat_list = [item for sublist in predictions for item in sublist]
df['prediction_2024'] = flat_list
df
```

```python
df['2024_ROC'] = (df['prediction_2024']-df['2019'])/df['2019']*100
df
```

```python
df['2024_Rank'] = df['2024_ROC'].rank(ascending = False)
df
```

```python
employment_prediction = df[['CBSA_Code','2024_ROC','2024_Rank']].copy()
employment_prediction
```
