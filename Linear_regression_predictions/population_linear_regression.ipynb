```python
import plotly.graph_objects as go
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
population_df = pd.read_csv('population_regions.csv')
population_df
```

```python
population_df.dtypes
```

```python
predictions = []
for i in range(0,384):
    x = [[2010],[2011],[2012],[2013],[2014],[2015],[2016],[2017],[2018],[2019]]
    y = population_df.iloc[i, 2:].values
    model.fit(x,y)
    y_pred = model.predict([[2024]])
    predictions.append(y_pred)
flat_list = [item for sublist in predictions for item in sublist]
population_df['prediction_2024'] = flat_list
population_df
```

```python
#calculate rate of change
population_df['prediction_2024'] = (((population_df['prediction_2024'] - population_df['2019']) / population_df['2019'])*100).round(1)
population_df
```

```python
population_df['Rank'] = population_df['prediction_2024'].rank(ascending=False)

population_df
```

```python
#drop columns
population_df = population_df.drop(columns = ['2010', '2011', '2012', '2013', '2014', 
                              '2015', '2016', '2017', '2018', '2019'])
population_df
```
