```python
import pandas as pd
import pymongo
from pymongo import MongoClient
```

```python
# Connection to mongo
client = MongoClient('mongodb+srv://<user>:<password>@cluster0.l3pqt.mongodb.net/MSA?retryWrites=true&w=majority')
# Select database
db = client['MSA']
# see list of collections
client.MSA.list_collection_names()
```

```python
# Select the collection with needed data
unemployment = db.unemployment_predicted_2024
dfu = pd.DataFrame(list(unemployment.find()))
print(dfu.dtypes)
dfu
```

```python
# change CBSA to object dtype
dfu['CBSA_Code'] = dfu['CBSA_Code'].astype('int')
# drop unnecessary columns
dfu.drop(columns=['_id', 'Metropolitan_Area'], inplace=True)
# rename columns
dfu.rename(columns={'CBSA_Code':'CBSA', 'Predicted_2024':'2024_Unem_ROC', 'Rank':'Unem_ROC_Rank'}, inplace=True)
print(dfu.dtypes)
dfu
```

```python
# Select the collection with needed data
employment = db.employment_prediction
dfe = pd.DataFrame(list(employment.find()))
print(dfe.dtypes)
dfe
```

```python
# change CBSA to object dtype
dfe['CBSA_Code'] = dfe['CBSA_Code'].astype('int')
# drop unnecessary columns
dfe.drop(columns='_id', inplace=True)
# rename columns
dfe.rename(columns={'CBSA_Code':'CBSA', '2024_ROC':'2024_Emp_ROC', '2024_Rank':'Emp_ROC_Rank'}, inplace=True)
dfe
```

```python
# Select the collection with needed data
gdp = db.GDP_predicted_2024
dfg = pd.DataFrame(list(gdp.find()))
print(dfg.dtypes)
dfg
```

```python
# drop unnecessary columns
dfg.drop(columns='_id', inplace=True)
# rename CBSA column to match other dataframes
dfg.rename(columns={'2024_ROC':'2024_GDP_ROC', '2024_Rank':'GDP_ROC_Rank'}, inplace=True)
print(dfg.dtypes)
dfg
```

```python
# Select the collection with needed data
population = db.population_predicted_2024
dfp = pd.DataFrame(list(population.find()))
print(dfp.dtypes)
dfp
```

```python
# drop unnecessary columns
dfp.drop(columns='_id', inplace=True)
# rename columns
dfp.rename(columns={'CBSA_Code':'CBSA', 'prediction_2024':'2024_Pop_ROC', 'Rank':'Pop_ROC_Rank'}, inplace=True)
print(dfp.dtypes)
dfp
```

```python
df_all_features = (dfp.merge(dfu, on='CBSA')
                   .merge(dfe, on='CBSA')
                   .merge(dfg, on='CBSA'))
df_all_features
```
