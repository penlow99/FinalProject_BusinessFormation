```python
# Select the collection with needed data
population = db.arima_pop_pred_score
dfp = pd.DataFrame(list(population.find()))
print(dfp.dtypes)
dfp
```

```python
# drop unnecessary columns
dfp.drop(columns=['_id'], inplace=True)
dfp.head()
```

```python
df_all_features = (dfp.merge(dfu, on='CBSA')
                   .merge(dfe, on='CBSA')
                   .merge(dfg, on='CBSA')
                   .merge(dfb, on='CBSA')
                   .merge(dfc, on='CBSA')
                   .merge(dfa, on='CBSA')
                   .merge(dfl, on='CBSA'))
df_all_features
```

```python
# reorder column names
df_all_features = df_all_features[['CBSA',
                                   'MSA',
                                   '2024_Pop_ROC',
                                   '2024_Unem_Rate',
                                   '2024_Emp_ROC',
                                   '2024_GDP_ROC',
                                   '2024_Bus_ROC',
                                   '2024_CPI',
                                   '2024_AMW_ROC',
                                   '2024_LP_ROC',
                                   'Pop_Score',
                                   'Unem_Score',
                                   'Emp_Score',
                                   'GDP_Score',
                                   'Bus_Score',
                                   'CPI_Score',
                                   'AMW_Score',
                                   'LP_Score']]
df_all_features
```

```python
df_rank_total = df_all_features[['MSA',
                                 'Pop_Score',
                                 'Unem_Score',
                                 'Emp_Score',
                                 'GDP_Score',
                                 'Bus_Score',
                                 'CPI_Score',
                                 'AMW_Score',
                                 'LP_Score']].copy()
df_rank_total
```

```python
df_rank_total['Total_Score'] = df_rank_total.sum(axis=1)
df_rank_total
```

```python
df_rank_total.sort_values(by='Total_Score', ascending=False).head(10)
```

```python
df_all_features['Total_Score'] = df_rank_total['Total_Score']
df_all_features
```

```python
df_all_features['Overall_Rank'] = df_all_features['Total_Score'].rank(ascending=False)
df_all_features
```

```python
df_all_features.to_csv('arima_2024_ALL_ROC_scores_total.csv', index=False)
```
